### Pearson Correlation Coefficient

Between -1 and 1

```python
def pearson_coeff(feat_1, feat_2, df):
    
    x = df[feat_1]
    y = df[feat_2]
    
    covar = np.cov(x,y)[0,1]
    std_x = np.std(x)
    std_y = np.std(y)
    
    
    print('Pearson Correlation Coefficient:')
    print(feat_1, "VS.", feat_2)
    
    return covar / (std_x * std_y)
```

```python
pearson_coeff(feat1, feat2, mi)
```
