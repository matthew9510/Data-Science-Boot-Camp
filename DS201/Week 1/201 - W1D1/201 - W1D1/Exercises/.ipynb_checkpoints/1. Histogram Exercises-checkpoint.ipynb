## Did the addition of the 3-Point Shot in 1979, affect the distribution of Total Points Scored?

I went ahead and made the new dataframe for everything before 1979, you make the one for everything after.

```python
no_three_point = nba[nba['Year'] < 1979]
three_point = nba[nba['Year'] > 1978]
```

### Preliminary Exploration

#### No 3-Pointer IN NBA

Use `.value_counts()` method on `PTS` column.

```python
no_three_point['PTS'].value_counts().sort_index(ascending=False).head()
```

Describe what the `value_counts()` method does.

Lists all unique values with a tally of how many times each unique values were in the dataset 

What is represented by the index?

The unique value of how many points were made. 

What is represented by the values?

The amount of times that value was found in the data set.

### Visual Exploration

#### No 3-Pointer IN NBA

```python
# use .hist() to plot the FG column
no_three_point['PTS'].hist(bins=20)

# Make it pretty
ax = plt.gca()
#ax.set_title("Points count Histogram") # more descriptive with how the two axes relate
ax.set_title("Total Points Scored By Players Before 1979")
ax.set_xlabel("Number Of Points")
ax.set_ylabel("Number Of Players")

# "print" the plot
plt.show()
```

#### 3-Pointer IN NBA

```python
# use .hist() to plot just the FG column for three_point
three_point['PTS'].hist(bins=20)
# Make it pretty
ax = plt.gca()
ax.set_title("Total Points Scored By Players After 1979")
ax.set_xlabel("Number Of Points")
ax.set_ylabel("Number Of Players")


# "print" the plot
plt.show()
```
