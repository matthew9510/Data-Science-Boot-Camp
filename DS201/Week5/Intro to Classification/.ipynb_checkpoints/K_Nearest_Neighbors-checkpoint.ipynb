# K Nearest Neighbors

### Imports

```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

### Load Data

```python
iris = datasets.load_iris()
```

```python
print(iris.feature_names)
```

## Build Model

### Assign X and y to hold data and target

X is capital because it is a matrix

```python
X = iris.data
y = iris.target
```

```python
X
```

```python
y
```

```python
X.shape
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
for row_ix, target in enumerate(y):
    print("**************")
    print("Example #" + str(row_ix))
    print("**************")
    for column_ix, column_value in enumerate(X[row_ix]):
        print(iris.feature_names[column_ix] + ":", column_value)
    print("-------------------")
    print("Target:", str(target) + " =", iris.target_names[target])
    print("-------------------")
    print()
    print()
```

### Split X and y into training and test sets

stratify - randomly chooses samples to split 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state = 2, stratify = y)
```

## Aside: Viualize data

```python
# Shows the different classes present in training data
plt.scatter(X_train[:,1],X_train[:,3], c=y_train)

plt.xlabel(iris["feature_names"][1])
plt.ylabel(iris["feature_names"][3])


# Shows the points that we are trying to classify in "blue", these
# are the points from the test set.
# plt.scatter(X_test[:,0], X_test[:,1], c='b')
plt.show()
```

### Create Model

```python
model = KNeighborsClassifier()
```

### Fit model to training data

```python
model.fit(X_train, y_train)
```

### Model Score

```python
#Higher score is good! :)
model.score(X_test, y_test)
```

### Make predictions!

```python
predictions = model.predict(X_test)
```

```python
y_test.shape
```

```python
predictions.shape
```

### "Eyeball" the accuracy of model.

```python
print("predictions:", predictions)
print("actual clss:", y_test)

#I adjusted the model to 5 nearest neighbors, now its a perfect match!
```

### Classification Report

```python
classification_report = metrics.classification_report(y_test, predictions)
print(classification_report)
```

### Confusion Matrix

```python
confusion_matrix = metrics.confusion_matrix(y_test, predictions)
print(confusion_matrix)
```
