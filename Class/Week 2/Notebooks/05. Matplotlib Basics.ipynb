## 1. Basic Matplotlib Commands

We can create a very simple line plots using the numpy arrays we created above.

### Single plot on a `figure` object

**Simple Plot**

```python
# Set a figure
fig = plt.figure()

# grab the `ax` object
ax = fig.gca()

# Plot data
ax.plot(x, y, 'r') # 'r' is the color red

# Clean up output
plt.show()
```

````python
# Set a figure 
fig = plt.figure()

# axis object
# grab the `ax` object
ax = fig.gca() # get current axes # put axis on figure; some sort of scales 

# Plot data
ax.plot(x, y, 'r') # 'r' is the color red
#ax.plot(x, x, 'b') # 'r' is the color red

plt.show()
````

**With Labels and Titles**

```python
# Set a figure
fig = plt.figure()

# Grab `ax` object
ax = fig.gca()

# Fig Title
fig.suptitle("Canvas Title")

# Ax Title
ax.set_title("Axis Title")

ax.plot(x, y, 'r') # 'r' is the color red

# Labels
ax.set_xlabel("X-label")
ax.set_ylabel("y-label")


# Clean up output
plt.show()
```

## 3. Titles, Axis Labels, and Legends

Now that we have covered the basics of how to create a figure canvas and add axes instances to the canvas, let's look at how decorate a figure with titles, axis labels, and legends.

<br>

### Figure titles

A title can be added to each axis instance in a figure. To set the title, use the `set_title` method in the axes instance:

```python
ax[0].set_title("Left")
ax[1].set_title("Right")
```

**In Context**

```python
# Instantiate figure and axis objects
fig, ax = plt.subplots(1,2)

# ax[0]
ax[0].set_title("Left")

# ax[1]
ax[1].set_title("Right")


# Clean up plt "output"
plt.show()
```

````python
# Instantiate figure and axis objects
fig, ax = plt.subplots(1,2)

# ax[0]
ax[0].set_title("Left")

# ax[1]
ax[1].set_title("Right")

print(type(ax[0]))
# outer-most title
fig.suptitle("Outer-most Title")

# Clean up plt "output"
plt.show()


# Instantiate figure and axis objects
fig, ax = plt.subplots(2,2)

# ax[row][column]
ax[0][0].set_title("00")
ax[0][1].set_title("01")
ax[1][0].set_title("10")
ax[1][1].set_title("11")

# outer-most title
fig.suptitle("Outer-most Title")

# Clean up plt "output"
plt.show()
````

<br>

### Axis labels

Similarly, with the methods `set_xlabel` and `set_ylabel`, we can set the labels of the X and Y axes:

```python
ax[0].set_xlabel("x",fontsize=30)
ax[0].set_ylabel("y", fontsize=30)

ax[1].set_xlabel("x",fontsize=20)
ax[1].set_ylabel("y", fontsize=20)
```

**In Context**

```python
# Instantiate figure and axes objects
fig, ax = plt.subplots(1,2)
# optional parameter: constrained_layout=True


# ax[0]
ax[0].set_title("Left")
ax[0].set_xlabel("x",fontsize=30)
ax[0].set_ylabel("y", fontsize=30)


# ax[1]
ax[1].set_title("Right")
ax[1].set_xlabel("x",fontsize=20)
ax[1].set_ylabel("y", fontsize=20)


# Clean-up plt "output"
plt.show()

```

````python
# Instantiate figure and axes objects
fig, ax = plt.subplots(1,2)
# optional parameter: constrained_layout=True


# ax[0]
ax[0].set_title("Left")
ax[0].set_xlabel("x",fontsize=30)
ax[0].set_ylabel("y", fontsize=30)
ax[0].plot(x,y)

# ax[1]
ax[1].set_title("Right")
ax[1].set_xlabel("y",fontsize=20)
ax[1].set_ylabel("x", fontsize=20)
ax[1].plot(y,x)


# Clean-up plt "output"
plt.show()
````

<br>

### Legends

You can use the **label="label text"** keyword argument when plots or other objects are added to the figure, and then using the **legend** method without arguments to add the legend to the figure: 

```python

fig = plt.figure(figsize=(12,4)) 

plt.plot(x, x**2, label="x^2")
plt.plot(x, x**3, label="x^3")

plt.legend(loc="upper left")

plt.show()

```

Use legend to show scale in terms of different graphs

````python
fig = plt.figure()
ax = fig.gca()

ax.plot(x, x**2, label="x^2")
ax.plot(x, x**3, label="x^3")

plt.legend(loc="upper left")

plt.show()
````

## 4. Setting `color`, `linewidth`, and `linetype`

Matplotlib gives you *a lot* of options for customizing colors, linewidths, and linetypes. 

There is the basic MATLAB like syntax (which I would suggest you avoid using for more clairty sake:

### Colors with the `color` parameter

We can also define colors by their names or RGB hex codes and optionally provide an alpha value using the `color` and `alpha` keyword arguments. Alpha indicates opacity.

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5) # half-transparant
ax.plot(x, x+2, color="#8B008B")        # RGB hex code
ax.plot(x, x+3, color="#FF8C00")        # RGB hex code 

plt.show()
```

````python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5) # half-transparant
ax.plot(x, x+2, color="#8B008B")        # RGB hex code
ax.plot(x, x+3, color="#FF8C00")        # RGB hex code 

plt.show()
````

### `line` and `marker` styles

To change the line width, we can use the `linewidth` or `lw` keyword argument. The line style can be selected using the `linestyle` or `ls` keyword arguments:

```python

fig, ax = plt.subplots(figsize=(12,6))

# linewidth
ax.plot(x, x+1, color= "red", linewidth= 0.25)
ax.plot(x, x+2, color= "red", lw= 0.50)
ax.plot(x, x+3, color= "red", lw= 1.00)
ax.plot(x, x+4, color= "red", lw= 2.00)

# possible linestype options -, , -., :, steps
ax.plot(x, x+5, color= "green", lw= 3, linestyle= '-')
ax.plot(x, x+6, color= "green", lw= 3, ls= '-.')
ax.plot(x, x+7, color= "green", lw= 3, ls= ':')

# custom dash
line, = ax.plot(x, x+8, color= "black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='x')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', ms=4, 
        markerfacecolor="white")
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', ms=8, 
        mfc="orange", markeredgewidth=3)
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', ms=8, 
        mfc="orange", mew=1, mec="green");

```

````python
fig, ax = plt.subplots(figsize=(12,6))

# linewidth
ax.plot(x, x+1, color= "red", linewidth= 0.25)
ax.plot(x, x+2, color= "red", lw= 0.50)
ax.plot(x, x+3, color= "red", lw= 1.00)
ax.plot(x, x+4, color= "red", lw= 2.00)

# possible linestype options -, , -., :, steps
ax.plot(x, x+5, color= "green", lw= 3, linestyle= '-')
ax.plot(x, x+6, color= "green", lw= 3, ls= '-.')
ax.plot(x, x+7, color= "green", lw= 3, ls= ':')

# custom dash
line, = ax.plot(x, x+8, color= "black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='x')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', ms=4, 
        markerfacecolor="white")
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', ms=8, 
        mfc="orange", markeredgewidth=3)
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', ms=8, 
        mfc="orange", mew=1, mec="green");
````

## 5. Control over axis appearance

**In this section we will look at controlling axis sizing properties in a matplotlib figure.**

### `set_ylim` and `set_xlim`

We can configure the ranges of the axes using the `set_ylim` and `set_xlim` methods in the axis object, or `axis('tight')` for automatically getting "tightly fitted" axes ranges:

```python

# Instantiate figure and axes objects
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

# ax[0]
axes[0].plot(x, x**2)
axes[0].plot(x, x**3)
axes[0].set_title("default axes ranges")



# ax[1]
axes[1].plot(x, x**2)
axes[1].plot(x, x**3)
axes[1].set_title("tight axes")

# set axis range with .axis() method
axes[1].axis('tight')



# ax[2]
axes[2].plot(x, x**2)
axes[2].plot(x, x**3)
axes[2].set_title("custom axes range");

# set axis range with `set_ylim()` and `set_xlim()` methods 
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])



plt.show()

```

````python
# Instantiate figure and axes objects
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

# ax[0]
axes[0].plot(x, x**2)
axes[0].plot(x, x**3)
axes[0].set_title("default axes ranges")



# ax[1]
axes[1].plot(x, x**2)
axes[1].plot(x, x**3)
axes[1].set_title("tight axes")

# set axis range with .axis() method
axes[1].axis('tight')



# ax[2]
axes[2].plot(x, x**2)
axes[2].plot(x, x**3)
axes[2].set_title("custom axes range");

# set axis range with `set_ylim()` and `set_xlim()` methods 
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])



plt.show()
````
